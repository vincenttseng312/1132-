```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# 建立模擬資料
data = {
    '公司代號': ['1101', '1102', '1103', '1104', '1105', '1106', '1107', '1108', '1109', '1110'],
    '流動比率': [1.25, 0.85, 1.10, 0.95, 1.40, 0.90, 1.30, 0.75, 1.05, 1.15],
    '負債比率': [55, 78, 62, 80, 40, 85, 50, 90, 60, 52],
    'ROA': [3.2, -1.1, 1.0, -2.5, 4.0, -1.8, 3.0, -3.0, 0.5, 2.0],
    'EPS': [1.8, -0.8, 0.3, -1.2, 2.2, -0.5, 1.5, -1.8, 0.0, 1.0],
    '營收年增率': [5.0, -12.0, -1.0, -20.0, 10.0, -15.0, 3.0, -25.0, 1.0, 2.5],
    '媒體情緒分數': [0.2, -0.6, -0.2, -0.9, 0.3, -0.7, 0.1, -1.0, -0.1, 0.0],
    '是否風險事件': [0, 1, 0, 1, 0, 1, 0, 1, 0, 0]
}

df = pd.DataFrame(data)

# 分出特徵與標籤
df_features = df[['公司代號', '流動比率', '負債比率', 'ROA', 'EPS', '營收年增率', '媒體情緒分數']]
X = df_features.drop(columns=['公司代號'])
y = df['是否風險事件']

# 分割訓練與測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
company_test = df_features.loc[X_test.index, '公司代號']

# 建立模型
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 預測
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]

# 顯示基本評估
print("混淆矩陣：")
print(confusion_matrix(y_test, y_pred))
print("\n分類報告：")
print(classification_report(y_test, y_pred))

# ----------- 圖1：風險機率分布 -----------
plt.figure(figsize=(8, 4))
sns.histplot(y_proba, bins=10, kde=True, color='teal')
plt.title("預測風險機率分布")
plt.xlabel("預測風險機率")
plt.ylabel("樣本數量")
plt.grid(True)
plt.tight_layout()
plt.show()

# ----------- 圖2：ROA 與風險事件關聯圖 -----------
plt.figure(figsize=(6, 4))
sns.boxplot(data=df, x='是否風險事件', y='ROA', palette='Set2')
plt.title("ROA 與風險事件的關聯")
plt.xticks([0, 1], ["無風險", "有風險"])
plt.tight_layout()
plt.show()

# ----------- 圖3：特徵重要性視覺化 -----------
importances = model.feature_importances_
feature_names = X.columns
feat_df = pd.DataFrame({'特徵': feature_names, '重要性': importances})
feat_df = feat_df.sort_values(by='重要性', ascending=False)

plt.figure(figsize=(8, 5))
sns.barplot(data=feat_df, x='重要性', y='特徵', palette='Blues_r')
plt.title("模型特徵重要性")
plt.tight_layout()
plt.show()

# ----------- 預測所有公司風險機率 -----------
all_X = df_features.drop(columns=['公司代號'])
all_proba = model.predict_proba(all_X)[:, 1]

# 整合結果
full_result_df = pd.DataFrame({
    '公司代號': df['公司代號'],
    '預測風險機率': all_proba,
    '實際是否風險事件': df['是否風險事件']
}).sort_values(by='預測風險機率', ascending=False)

print("\n📋 所有公司風險預測結果：")
print(full_result_df.to_string(index=False))
```
